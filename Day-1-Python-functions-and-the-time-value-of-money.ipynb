## Task 1: Functions for time value of money calculations
The **time value of money** is a fundamental concept in finance that refers to the idea that a dollar received today is worth more than a dollar received in the future. This is because money today can be invested and earn interest over time, making it more valuable in the future. 



The equation for the future value of a lump sum: $$FV = PV(1+r)^N$$
where FV: future value,\
PV: present value,\
r: interest rate, \
N: the number of periods. 

Write a function fv_lump_sum(r, n, pv) to calculate and return the future value of a lump sum of present value invested at the periodic rate r for n periods. 

```python
def fv_lump_sum(rate, period, pv):
    fv=pv*(1+rate)**period
    return f"${fv:.2f}"
```

```python
# $100 at 5% rate for 2 years
fv_lump_sum(0.05,2,100)
```

```python
# $400 at 8% APR for 20 years (with monthly compounding)
fv_lump_sum(0.08/12, 20*12, 400)
```

We can calculate the present value (PV) by discounting the future value (FV) at a given rate and time. 
The equation for the present value of a lump sum:

$$PV = \frac{FV}{(1+r)^N}$$
where r: discount rate,\
N: number of periods. 

This formula assumes that the interest rate is constant over time, which is not always the case in practice. 

Write a function pv_lump_sum(r, n, fv) to calculate and return the present value of a lump sum fv to be received in the future, discounted at the periodic rate r for n periods. 

```python
def pv_lump_sum(rate, period, fv):
    pv = fv/((1+rate)**period)
    return f'${pv:.2f}'
```

```python
# $1000 to be received in 5 years at 6% per year
pv_lump_sum(0.06,5,1000)
```

```python
# $500 received in 5 years, 6% APR, semi-annually compounding
pv_lump_sum(0.06/2,5*2,500)
```

**Annuity** is stream of *equal cash flows* that occurs at *equal intervals* over a *finite period*. Annuity calculation are normally used in fixed income (fixed rate), e.g. Bonds. A bond is a debt security issued by a borrower to raise capital. When you purchase a bond, and the lender lend money to the issuer, who promises to pay back the principal amount plus interest at a specified date in the future.

There are two types of annuities: 
1. Ordinary annuities - the periodic payments are made at the end of each period 
2. Annuities due - the periodic payments are made at the beginning of each period 

The equation for the future value of an **ordinary annuity** is:

$$PV = PMT\frac{1-(1+r)^{-n}}{r}$$

where PV: present value,\
PMT: periodic payment or cash flow,\
r: interest rate per period,\
n: number of periods.

This equation calculates the future value of a series of equal periodic payments (annuity) by discounting each payment back to the present value at the given interest rate. The formula takes into account the compounding effect of the interest rate over the specified number of periods.

It's important to note that this formula assumes the periodic payments are made at the end of each period and that the interest rate remains constant throughout the duration of the annuity.

There are several occassions of ordinary annuity:
- Retiree receives monthly payments of from the pension plan. The payments will continue for the rest of her life. 
- Loan repayment. A borrower who makes monthly payments on a loan is making an ordinary annuity payment to the lender.
- Rent payments and the landlord receives a fixed amount of money at the end of each month for a fixed number of months.


Write a function pv_ordinary_annuity(r, n, pmt) to calculate and return the present value of an annuity of pmt to be received each period for n periods, at the periodic rate r. 


```python
def pv_ordinary_annuity(rate, payment_freq, payment_amount):
    pv = payment_amount*((1-(1+rate)**(-payment_freq))/rate)
    return f"${pv:.2f}"
```

```python
# pv of 30 payments of $250 per year, 5% interest
pv_ordinary_annuity(0.05, 30, 250)
```

```python
# pv of 60 payments of $471.75 per month, 0.9% APR
pv_ordinary_annuity(0.009/12, 60, 471.75)
```

The equation for the future value of an **annuity due** is:

$$PV = PMT\frac{1-(1+r)^{-n}}{r} \times (1+r)$$

where PV: present value,\
PMT: periodic payment or cash flow,\
r: interest rate per period,\
n: number of periods.

An example of an annuity due is a lease payment or rent paid at the beginning of each month.

Write a function fv_annuity(r, n, pmt) to calculate and return the future value of an annuity of pmt to be received each period for n periods, invested at the periodic rate r. 

```python
def pv_annuity_due(rate, payment_freq, payment_amount):
    pv = payment_amount*((1-(1+rate)**(-payment_freq))/rate)*(1+rate)
    return f"${pv:.2f}"
```

```python
# pv of 30 payments of $250 per year, 5% interest
pv_annuity_due(0.05,30,250)
# note that the annuity due pv > ordinary annuity pv
```

```python
# pv of 60 payments of $471.75 per month, 0.9% APR
pv_annuity_due(0.009/12, 60, 471.75)
```


The equation for the amortizing annuity payment is used to calculate the fixed payment that should be made on a loan or mortgage in order to pay off the debt over a specific period of time, while also paying interest on the outstanding balance assuming the rate is fixed during this period of time.

The equation for the **amortizing annuity payment** is:

$$PMT = \frac{r*PV}{1-(1+r)^{-n}}$$

where PV: present value,\
PMT: periodic payment or cash flow,\
r: interest rate per period,\
n: number of periods.

Write a function annuity_payment(r, n, pv) that calculates the amortizing annuity
payment for a present value of pv to be repaid at a periodic interest rate of r for n periods.



```python
def annuity_payment(rate, year, present_value):
    pmt = (rate*present_value)/(1-(1+rate)**(-year))
    return f"${pmt:.2f}"
```

```python
# annuity payment for pv of $1,000 for 10 year at 5%
annuity_payment(0.05, 10, 1000)
```

```python
# annuity payment for pv of $27,667 for 60 months at 0.9% APR
annuity_payment(0.009/12, 60, 27667.44)
```
